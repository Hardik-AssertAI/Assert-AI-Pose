```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import GridSearchCV
import os
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
import pickle
```

```python
df = pd.read_csv('facial_expressions.csv')
df
```

```python
y=df['Y']
x=df.drop(['Image_name', 'Y'], axis = 1)
classes = {"Y": {"angry": 0, "happy": 1, "sad": 2}}
df = df.replace(classes)
y=df['Y']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=77,stratify=y)
print('Splitted Successfully')
```

```python
# Standardize input 

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train_standardized = sc_X.fit_transform(x_train)
x_test_standardized = sc_X.transform(x_test)
```

```python
# Dont forget to dump the scaler to use on new data
            # VERY IMPORTANT #

pickle.dump(sc_X, open('scaler_64_final.pkl','wb')) # change name to scaler_imagesize
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.01,0.1],'kernel':['linear']}
svc=svm.SVC(probability=True)
print("The training of the model is started, please wait for while as it may take few minutes to complete")
model=GridSearchCV(svc,param_grid)
model.fit(x_train_standardized,y_train)
print('The Model is trained well with the given images')
model.best_params_
```

```python
y_pred=model.predict(x_test_standardized)
print("The predicted Data is :")
y_pred
```

```python
print("The actual data is:")
print(np.array(y_test))

classification_report(y_pred,y_test)
print(f"The model is {accuracy_score(y_pred,y_test)*100}% accurate")
confusion_matrix(y_pred,y_test)
```

```python
pickle.dump(model,open('img_model_64_final.p','wb'))
print("Pickle is dumped successfully")
```
